```python
#Bài 2
import os
import math

def GetWinner(board):
    """Kiểm tra xem ai là người chiến thắng."""
    # Các tổ hợp thắng (hàng, cột, chéo)
    win_combinations = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8], # Hàng
        [0, 3, 6], [1, 4, 7], [2, 5, 8], # Cột
        [0, 4, 8], [2, 4, 6]             # Chéo
    ]
    for combo in win_combinations:
        if board[combo[0]] == board[combo[1]] == board[combo[2]]:
            return board[combo[0]]
    return None

def GetAvailableCells(board):
    """Lấy danh sách các ô còn trống."""
    return [cell for cell in board if isinstance(cell, int)]

def PrintBoard(board):
    """Hiển thị bảng cờ lên console."""
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f"\n {board[0]} | {board[1]} | {board[2]} ")
    print("-----------")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("-----------")
    print(f" {board[6]} | {board[7]} | {board[8]} \n")

def minimax(board, depth, alpha, beta, isMaximizing):
    """Thuật toán Minimax kết hợp cắt tỉa Alpha-Beta."""
    winner = GetWinner(board)
    if winner == "X": return 10 - depth
    if winner == "O": return -10 + depth
    if not GetAvailableCells(board): return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(board):
            board[cell - 1] = "X"
            eval = minimax(board, depth + 1, alpha, beta, False)
            board[cell - 1] = cell # Quay lui (Backtracking)
            maxEval = max(maxEval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break # Cắt tỉa nhánh Minimizing
        return maxEval
    else:
        minEval = math.inf
        for cell in GetAvailableCells(board):
            board[cell - 1] = "O"
            eval = minimax(board, depth + 1, alpha, beta, True)
            board[cell - 1] = cell # Quay lui
            minEval = min(minEval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break # Cắt tỉa nhánh Maximizing
        return minEval

def FindBestMove(board, AI_player):
    """Tìm nước đi tối ưu cho AI."""
    bestVal = -math.inf if AI_player == "X" else math.inf
    bestMove = -1
    isMaximizing = (AI_player == "X")

    for cell in GetAvailableCells(board):
        board[cell - 1] = AI_player
        # Sau khi AI đi, lượt tiếp theo sẽ là đối thủ (ngược lại của isMaximizing)
        moveVal = minimax(board, 0, -math.inf, math.inf, not isMaximizing)
        board[cell - 1] = cell

        if AI_player == "X":
            if moveVal > bestVal:
                bestVal, bestMove = moveVal, cell
        else:
            if moveVal < bestVal:
                bestVal, bestMove = moveVal, cell
    return bestMove

def main():
    player = input("Chọn quân cờ của bạn (X/O): ").strip().upper()
    while player not in ["X", "O"]:
        player = input("Vui lòng chỉ chọn X hoặc O: ").strip().upper()

    AI = "O" if player == "X" else "X"
    board = [*range(1, 10)]
    currentTurn = "X" # X luôn đi trước

    while True:
        PrintBoard(board)
        winner = GetWinner(board)
        if winner:
            print(f"KẾT THÚC: {winner} CHIẾN THẮNG!")
            break
        if not GetAvailableCells(board):
            print("KẾT THÚC: HÒA!")
            break

        if currentTurn == player:
            try:
                move = int(input(f"Lượt của bạn ({player}). Nhập số ô (1-9): "))
                if move in GetAvailableCells(board):
                    board[move - 1] = player
                    currentTurn = AI
                else:
                    print("Ô không hợp lệ hoặc đã có người đi!")
                    continue
            except ValueError:
                print("Vui lòng nhập số từ 1 đến 9.")
        else:
            print(f"AI ({AI}) đang tính toán...")
            move = FindBestMove(board, AI)
            board[move - 1] = AI
            currentTurn = player

if __name__ == "__main__":
    main()
```
