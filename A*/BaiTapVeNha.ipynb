```python
import heapq

class TSP_AStar:
    def __init__(self, matrix, city_names):
        self.matrix = matrix  # Ma trận kề chi phí giữa các thành phố
        self.n = len(matrix)
        self.city_names = city_names

    def h(self, current_city, unvisited_cities):
        """
        Hàm Heuristic h(n): Ước lượng chi phí từ đỉnh hiện tại đi qua
        các đỉnh còn lại. Một cách đơn giản là lấy số lượng đỉnh còn lại
        nhân với cạnh nhỏ nhất của đồ thị.
        """
        if not unvisited_cities:
            # Nếu không còn đỉnh nào, chi phí là quãng đường quay về đỉnh gốc (0)
            return self.matrix[current_city][0]

        # Heuristic đơn giản: Số thành phố còn lại * cạnh nhỏ nhất trong đồ thị
        return len(unvisited_cities) * min([min(row) for row in self.matrix if max(row) > 0])

    def solve(self, start_node_idx=0):
        # Mỗi node trong Priority Queue: (f_score, current_city, visited_tuple, g_score)
        # visited_tuple lưu danh sách các thành phố đã qua để hash được
        start_visited = tuple([start_node_idx])
        initial_h = self.h(start_node_idx, set(range(self.n)) - set(start_visited))

        # (f, current_city, visited, g)
        pq = [(initial_h, start_node_idx, start_visited, 0)]

        while pq:
            f, u, visited, g = heapq.heappop(pq)

            # Nếu đã đi qua tất cả các thành phố
            if len(visited) == self.n:
                # Cộng thêm chi phí quay về thành phố xuất phát
                total_cost = g + self.matrix[u][start_node_idx]
                full_path = [self.city_names[i] for i in visited] + [self.city_names[start_node_idx]]
                print(f"Lộ trình: {' -> '.join(full_path)}")
                print(f"Tổng chi phí: {total_cost}")
                return full_path, total_cost

            # Duyệt các thành phố chưa đi qua
            for v in range(self.n):
                if v not in visited:
                    new_g = g + self.matrix[u][v]
                    new_visited = visited + (v,)
                    new_h = self.h(v, set(range(self.n)) - set(new_visited))
                    new_f = new_g + new_h

                    heapq.heappush(pq, (new_f, v, new_visited, new_g))

        return None

# --- Thử nghiệm ---

# Tên các thành phố
cities = ['Hà Nội', 'Hải Phòng', 'Đà Nẵng', 'TP.HCM']

# Ma trận chi phí (0 nghĩa là cùng một thành phố)
# Ví dụ: matrix[0][1] là chi phí từ Hà Nội đi Hải Phòng
cost_matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

tsp_solver = TSP_AStar(cost_matrix, cities)
tsp_solver.solve()
```
