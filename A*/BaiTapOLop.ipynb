```python
from collections import deque

class Graph:
    def __init__(self, adjac_list, h_map):
        self.adjac_list = adjac_list
        self.h_map = h_map

    def get_neighbors(self, v):
        return self.adjac_list.get(v, [])

    def h(self, n):
        return self.h_map.get(n, float('inf'))

    def a_star_algorithm(self, start, stop):
        open_lst = set([start])
        closed_lst = set([])
        g_score = {}
        g_score[start] = 0
        par = {}
        par[start] = start

        while len(open_lst) > 0:
            n = None
            for v in open_lst:
                if n == None or g_score[v] + self.h(v) < g_score[n] + self.h(n):
                    n = v

            if n == None:
                print('Không tìm thấy đường đi!')
                return None
            if n == stop:
                reconst_path = []
                while par[n] != n:
                    reconst_path.append(n)
                    n = par[n]
                reconst_path.append(start)
                reconst_path.reverse()

                print(f"Đường đi tìm thấy: {' -> '.join(reconst_path)}")
                print(f"Tổng chi phí: {g_score[stop]}")
                return reconst_path
            for (m, weight) in self.get_neighbors(n):
                tentative_g_score = g_score[n] + weight

                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    par[m] = n
                    g_score[m] = tentative_g_score

                elif tentative_g_score < g_score.get(m, float('inf')):
                    g_score[m] = tentative_g_score
                    par[m] = n

                    if m in closed_lst:
                        closed_lst.remove(m)
                        open_lst.add(m)

            open_lst.remove(n)
            closed_lst.add(n)

        print('Không tồn tại đường đi giữa hai đỉnh này!')
        return None

adj_list = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('C', 1), ('D', 5)],
    'C': [('D', 2)],
    'D': []
}


heuristic_map = {
    'A': 4,
    'B': 2,
    'C': 1,
    'D': 0
}

graph = Graph(adj_list, heuristic_map)
print("Đang tìm đường từ A đến D")
graph.a_star_algorithm('A', 'D')
```
