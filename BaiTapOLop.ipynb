```python
# 13
def tron_mang(a, b):
    res = []
    length = max(len(a), len(b))
    for i in range(length):
        val_a = a[i] if i < len(a) else 0
        val_b = b[i] if i < len(b) else 0
        res.append(val_a + val_b)
    return res

mang_a = [3, 9, 1, 4]
mang_b = [2, 7, 4, 3, 2, 8]
print(f"Kết quả trộn: {tron_mang(mang_a, mang_b)}")
import random
```

```python
import random
import math

# a. Tạo mảng a chứa các số nguyên ngẫu nhiên
def tao_mang(m, n):
    # Tạo giá trị từ 0-20 để tổng dòng dễ nằm trong khoảng 45 cho bạn quan sát
    return [[random.randint(0, 20) for _ in range(n)] for _ in range(m)]

# b. Xuất các phần tử thuộc dòng k
def xuat_dong_k(a, k):
    if 0 <= k < len(a):
        print(f"Dòng {k}: {a[k]}")
    else:
        print(f"Dòng {k} không tồn tại.")

# c. Xuất các phần tử thuộc cột k
def xuat_cot_k(a, m, k):
    if 0 <= k < len(a[0]):
        cot = [a[i][k] for i in range(m)]
        print(f"Cột {k}: {cot}")
    else:
        print(f"Cột {k} không tồn tại.")

# d. Tìm dòng có tổng lớn nhất (Ưu tiên dòng có tổng <= 45)
def dong_tong_max_limit(a):
    max_tong = -1
    idx = -1
    for i, dong in enumerate(a):
        tong = sum(dong)
        if tong <= 45 and tong > max_tong:
            max_tong = tong
            idx = i

    # Nếu không có dòng nào <= 45, tìm dòng có tổng lớn nhất tuyệt đối
    if idx == -1:
        sums = [sum(d) for d in a]
        max_val = max(sums)
        idx = sums.index(max_val)
        print(f"(Lưu ý: Không có dòng nào <= 45, lấy dòng có tổng lớn nhất là {max_val})")

    return idx

# e. Tìm cột có tích nhỏ nhất
def cot_tich_min(a, m, n):
    min_tich = float('inf')
    idx = 0
    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]
        if tich < min_tich:
            min_tich = tich
            idx = j
    return idx

# f. Xuất các phần tử thuộc dòng chẵn và cột lẻ
def xuat_dong_chan_cot_le(a, m, n):
    res = [a[i][j] for i in range(0, m, 2) for j in range(1, n, 2)]
    print("Các phần tử dòng chẵn (0,2...), cột lẻ (1,3...):", res)

# g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ
def tbc_chan_dong_le(a, m, n):
    ds = [a[i][j] for i in range(1, m, 2) for j in range(n) if a[i][j] % 2 == 0]
    return sum(ds) / len(ds) if ds else 0

# h. Tính trung bình cộng các phần tử thuộc biên
def tbc_bien(a, m, n):
    bien = []
    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                bien.append(a[i][j])
    return sum(bien) / len(bien) if bien else 0

# i. Tính trung bình tích (Trung bình nhân) các phần tử không thuộc biên
def trung_binh_tich_noi_bo(a, m, n):
    noi_bo = [a[i][j] for i in range(1, m-1) for j in range(1, n-1)]
    if not noi_bo: return 0
    tich = 1
    for x in noi_bo: tich *= x
    # Căn bậc n của tích
    return tich ** (1/len(noi_bo))

m, n = 4, 4
matrix = tao_mang(m, n)

print("--- MA TRẬN VỪA TẠO ---")
for row in matrix:
    print(row)
print("-" * 30)

# Gọi các hàm để kiểm tra kết quả
xuat_dong_k(matrix, 1)
xuat_cot_k(matrix, m, 2)

idx_d = dong_tong_max_limit(matrix)
print(f"d. Dòng thỏa mãn điều kiện tổng lớn nhất: {idx_d} (Tổng = {sum(matrix[idx_d])})")

print(f"e. Chỉ số cột có tích nhỏ nhất: {cot_tich_min(matrix, m, n)}")

xuat_dong_chan_cot_le(matrix, m, n)

print(f"g. TBC phần tử chẵn thuộc dòng lẻ: {tbc_chan_dong_le(matrix, m, n):.2f}")
print(f"h. TBC các phần tử biên: {tbc_bien(matrix, m, n):.2f}")
print(f"i. Trung bình tích phần tử bên trong: {trung_binh_tich_noi_bo(matrix, m, n):.2f}")
```

```python
#15
# a. Xây dựng cấu trúc SINHVIEN
class SinhVien:
    def __init__(self, mssv, ten, nam_sinh, dtb):
        self.mssv = str(mssv).strip()
        self.ten = str(ten).strip()
        self.nam_sinh = int(nam_sinh)
        self.dtb = float(dtb)

# b. Tạo danh sách sinh viên mẫu
danh_sach_sv = [
    SinhVien("01DH0001", "Phan Anh Lan", 2004, 7.5),
    SinhVien("02DH0002", "Nguyen Lan", 2006, 4.5),
    SinhVien("03CD0001", "Phan Van Tu", 2002, 8.0),
    SinhVien("04DH0005", "Tran Thi Lan", 2006, 9.0)
]

# c. Đếm sinh viên đủ điều kiện lên lớp (DTB >= 5)
def sv_len_lop(ds):
    res = [sv for sv in ds if sv.dtb >= 5]
    return len(res)

# d. Xuất sinh viên đủ 20 tuổi (Tính theo năm hiện tại 2026)
def xuat_sv_20_tuoi(ds):
    nam_hien_tai = 2026
    found = False
    print("--- Danh sách sinh viên đủ 20 tuổi ---")
    for sv in ds:
        if nam_hien_tai - sv.nam_sinh == 20:
            print(f"- {sv.ten} ({sv.nam_sinh})")
            found = True
    if not found: print("Không có sinh viên nào 20 tuổi.")

# e. Đếm sinh viên hệ đại học (Ký tự 2,3 là 'DH' -> index 1,2)
def dem_sv_dai_hoc(ds):
    count = 0
    for sv in ds:
        # Dùng .upper() để so sánh chính xác hơn
        if len(sv.mssv) >= 3 and sv.mssv[1:3].upper() == "DH":
            count += 1
    return count

# f. Đếm sinh viên có tên "Lan"
def dem_ten_lan(ds):
    count = 0
    for sv in ds:
        parts = sv.ten.split()
        if parts and parts[-1].lower() == "lan":
            count += 1
    return count

# g. Đếm sinh viên có họ "Phan"
def dem_ho_phan(ds):
    count = 0
    for sv in ds:
        parts = sv.ten.split()
        if parts and parts[0].lower() == "phan":
            count += 1
    return count

print(f"1. Số SV đủ điều kiện lên lớp: {sv_len_lop(danh_sach_sv)}")
print(f"2. Số SV hệ Đại học: {dem_sv_dai_hoc(danh_sach_sv)}")
print(f"3. Số SV tên 'Lan': {dem_ten_lan(danh_sach_sv)}")
print(f"4. Số SV họ 'Phan': {dem_ho_phan(danh_sach_sv)}")
xuat_sv_20_tuoi(danh_sach_sv)
```
