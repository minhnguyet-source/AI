```python
#1
def tron_mang_min(a, b):
    # Lấy độ dài ngắn nhất để so sánh tương ứng
    n_min = min(len(a), len(b))

    # So sánh các phần tử tương ứng
    res = [min(a[i], b[i]) for i in range(n_min)]

    # Nối phần dư của mảng dài hơn (nếu có)
    if len(a) > n_min:
        res.extend(a[n_min:])
    elif len(b) > n_min:
        res.extend(b[n_min:])

    return res

# Chạy thử ví dụ của đề bài
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print(f"Mảng kết quả: {tron_mang_min(a, b)}")
```

```python
#2
def tong_tam_giac_tren_phu(a, n):
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong

def tri_tuyet_doi(a, m, n):
    for i in range(m):
        for j in range(n):
            a[i][j] = abs(a[i][j])

def thay_chan_bang_x(a, m, n, x):
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 == 0:
                a[i][j] = x

def kiem_tra_toan_chan(a, m, n):
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 != 0:
                return False
    return True

def la_doi_xung(a, n):
    for i in range(n):
        for j in range(i + 1, n):
            if a[i][j] != a[j][i]:
                return False
    return True

def duong_cheo_chinh_tang(a, n):
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True

def xuat_tam_giac_duoi_phu(a, n):
    print("Tam giác dưới đường chéo phụ:")
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end="\t")
        print()

def duong_cheo_phu_giam(a, n):
    for i in range(n - 1):
        if a[i][n-1-i] <= a[i+1][n-2-i]:
            return False
    return True

# 1. Tạo ma trận mẫu cấp 3 (3x3)
n = 3
m = 3
ma_tran = [
    [1, -2, 3],
    [2, 5, -4],
    [3, 4, 9]
]

print("Ma trận ban đầu:")
for row in ma_tran: print(row)

# 2. Gọi thử các hàm
print(f"\na. Tổng tam giác trên đường chéo phụ: {tong_tam_giac_tren_phu(ma_tran, n)}")
print(f"d. Ma trận toàn chẵn? {kiem_tra_toan_chan(ma_tran, m, n)}")
print(f"e. Ma trận đối xứng? {la_doi_xung(ma_tran, n)}")
print(f"f. Đường chéo chính tăng dần? {duong_cheo_chinh_tang(ma_tran, n)}")
print(f"h. Đường chéo phụ giảm dần? {duong_cheo_phu_giam(ma_tran, n)}")

print("\ng. Xuất tam giác dưới phụ:")
xuat_tam_giac_duoi_phu(ma_tran, n)

# 3. Thử hàm thay đổi giá trị
thay_chan_bang_x(ma_tran, m, n, 99)
print("\nMa trận sau khi thay số chẵn bằng 99:")
for row in ma_tran: print(row)
```
