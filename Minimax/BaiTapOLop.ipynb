```python
#Bài 1
import copy
import math
import numpy

X = "X"
O = "O"
EMPTY = None

def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    """Xác định lượt của ai dựa trên số lượng ký tự trên bảng."""
    count_x = sum(row.count(X) for row in board)
    count_o = sum(row.count(O) for row in board)
    return X if count_x <= count_o else O

def actions(board):
    """Trả về tập hợp các tọa độ (i, j) còn trống."""
    return {(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY}

def result(board, action):
    """Tạo bảng mới sau khi thực hiện nước đi."""
    curr_player = player(board)
    new_board = copy.deepcopy(board)
    i, j = action
    new_board[i][j] = curr_player
    return new_board

def winner(board):
    """Kiểm tra người thắng cuộc."""
    # Kiểm tra hàng và cột
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] and board[i][0] is not None:
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] and board[0][i] is not None:
            return board[0][i]
    # Kiểm tra đường chéo
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] is not None:
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] is not None:
        return board[0][2]
    return None

def terminal(board):
    """Kiểm tra trò chơi kết thúc chưa."""
    if winner(board) or not actions(board):
        return True
    return False

def utility(board):
    """Trả về điểm số: X thắng = 1, O thắng = -1, Hòa = 0."""
    res = winner(board)
    if res == X: return 1
    if res == O: return -1
    return 0

# --- THUẬT TOÁN MINIMAX ---

def maxValue(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v

def minimax(board):
    """Tìm nước đi tối ưu nhất cho người chơi hiện tại."""
    curr_player = player(board)
    best_move = None

    if curr_player == X:
        best_v = -math.inf
        for action in actions(board):
            v = minValue(result(board, action))
            if v > best_v:
                best_v = v
                best_move = action
    else:
        best_v = math.inf
        for action in actions(board):
            v = maxValue(result(board, action))
            if v < best_v:
                best_v = v
                best_move = action
    return best_move

# --- LUỒNG CHÍNH ---

if __name__ == "__main__":
    board = initial_state()
    print("--- CHÀO MỪNG ĐẾN VỚI TIC-TAC-TOE MINIMAX ---")
    user_choice = input("Bạn muốn chơi quân gì? (X đi trước): ").upper()
    user = X if user_choice == X else O
    ai = O if user == X else X

    while not terminal(board):
        curr_p = player(board)
        print(f"\nLượt của: {curr_p}")
        print(numpy.array(board))

        if curr_p == user:
            try:
                row = int(input("Nhập hàng (0-2): "))
                col = int(input("Nhập cột (0-2): "))
                if (row, col) in actions(board):
                    board[row][col] = user
                else:
                    print("Vị trí không hợp lệ!")
                    continue
            except ValueError:
                print("Vui lòng nhập số!")
        else:
            print("AI đang suy nghĩ...")
            move = minimax(board)
            board = result(board, move)

    # Kết quả chung cuộc
    print("\n--- KẾT THÚC ---")
    print(numpy.array(board))
    res = winner(board)
    if res:
        print(f"Người chiến thắng là: {res}")
    else:
        print("Trò chơi Hòa!")
     2
```
